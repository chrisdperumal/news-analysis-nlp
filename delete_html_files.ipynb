```python
import os

def remove_html_files(directory):
    # Walk through all subdirectories and files in 'directory'
    for root, dirs, files in os.walk(directory):
        for file in files:
            # Check if the file ends with .html
            if file.endswith('.html'):
                # Construct full file path
                file_path = os.path.join(root, file)
                # Remove the file
                os.remove(file_path)
                print(f"Removed: {file_path}")
```

```python
remove_html_files("/Users/chrisperumal/news-please-repo/data/2024/04/17")
```
